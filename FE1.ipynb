```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
from skimage.feature import greycomatrix, greycoprops

# Load the live and dead cattle muzzle images
live_img = cv2.imread('/content/20230408_055504.jpg')
dead_img = cv2.imread('/content/20230408_070134.jpg')

# Step 1: Color Features

# Calculate average RGB values for live and dead images
live_avg_color = np.mean(live_img, axis=(0, 1))
dead_avg_color = np.mean(dead_img, axis=(0, 1))

# Display average RGB values
print("Live Average Color (RGB):", live_avg_color)
print("Dead Average Color (RGB):", dead_avg_color)

# Step 2: Texture Features

# Convert images to grayscale
live_gray = cv2.cvtColor(live_img, cv2.COLOR_BGR2GRAY)
dead_gray = cv2.cvtColor(dead_img, cv2.COLOR_BGR2GRAY)

# Compute local binary patterns (LBP)
live_lbp = local_binary_pattern(live_gray, 8, 1, method='uniform')
dead_lbp = local_binary_pattern(dead_gray, 8, 1, method='uniform')

# Calculate LBP histograms
live_hist, _ = np.histogram(live_lbp.ravel(), bins=np.arange(0, 10))
dead_hist, _ = np.histogram(dead_lbp.ravel(), bins=np.arange(0, 10))

# Display LBP histograms
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(np.arange(9), live_hist)
plt.title('Live Muzzle LBP Histogram')
plt.xlabel('LBP Bin')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.bar(np.arange(9), dead_hist)
plt.title('Dead Muzzle LBP Histogram')
plt.xlabel('LBP Bin')
plt.ylabel('Frequency')
plt.show()

# Step 3: Shape Features

# Convert images to grayscale and apply edge detection
live_gray = cv2.cvtColor(live_img, cv2.COLOR_BGR2GRAY)
dead_gray = cv2.cvtColor(dead_img, cv2.COLOR_BGR2GRAY)
live_edges = cv2.Canny(live_gray, 50, 150)
dead_edges = cv2.Canny(dead_gray, 50, 150)

# Calculate contour areas
live_contour, _ = cv2.findContours(live_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
dead_contour, _ = cv2.findContours(dead_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
live_area = cv2.contourArea(max(live_contour, key=cv2.contourArea))
dead_area = cv2.contourArea(max(dead_contour, key=cv2.contourArea))

# Display contour areas
print("Live Muzzle Area:", live_area)
print("Dead Muzzle Area:", dead_area)

# Step 4: Motion Features (if applicable)

# If you have a sequence of images showing motion, you can calculate motion features such as optical flow
# However, as
# Step 4: Motion Features (not applicable for static images)

# Since we are working with static images, motion features are not applicable.

# Display the images and extracted features
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(live_img, cv2.COLOR_BGR2RGB))
plt.title('Live Muzzle Image')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(dead_img, cv2.COLOR_BGR2RGB))
plt.title('Dead Muzzle Image')
plt.axis('off')

# Step 1: Color Features
plt.subplot(2, 3, 4)
plt.imshow([live_avg_color.astype(int)], aspect='auto')
plt.title('Live Muzzle Average Color (RGB)')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow([dead_avg_color.astype(int)], aspect='auto')
plt.title('Dead Muzzle Average Color (RGB)')
plt.axis('off')

# Step 2: Texture Features
plt.subplot(2, 3, 3)
plt.imshow(live_gray, cmap='gray')
plt.title('Live Muzzle (Grayscale)')
plt.axis('off')
plt.subplot(2, 3, 6)
plt.imshow(dead_gray, cmap='gray')
plt.title('Dead Muzzle (Grayscale)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
